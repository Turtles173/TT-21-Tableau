### This is the data cleanup file to merge the monthly data files from the NYC citibike program and the COVID cases since January 2020 to examine the effect of COVID on the number of trips taken over that period.

```python
# Dependancies
import pandas as pd
import numpy as np
import os
```

### Clean up and concat the citibike csv's

```python
# Bike data downloaded from https://www.citibikenyc.com/system-data from Jan 2020 to August 2020
bike_jan_path = "data/JC-202001-citibike-tripdata.csv"
bike_feb_path = "data/JC-202002-citibike-tripdata.csv"
bike_mar_path = "data/JC-202003-citibike-tripdata.csv"
bike_apr_path = "data/JC-202004-citibike-tripdata.csv"
bike_may_path = "data/JC-202005-citibike-tripdata.csv"
bike_jun_path = "data/JC-202006-citibike-tripdata.csv"
bike_jul_path = "data/JC-202007-citibike-tripdata.csv"
bike_aug_path = "data/JC-202008-citibike-tripdata.csv"
```

```python
# Read in the bike data
bike_jan = pd.read_csv(bike_jan_path)
bike_feb = pd.read_csv(bike_feb_path)
bike_mar = pd.read_csv(bike_mar_path)
bike_apr = pd.read_csv(bike_apr_path)
bike_may = pd.read_csv(bike_may_path)
bike_jun = pd.read_csv(bike_jun_path)
bike_jul = pd.read_csv(bike_jul_path)
bike_aug = pd.read_csv(bike_aug_path)
```

```python
# Check the data columns and import
bike_aug.head()
```

```python
# Concat the files to create a full dataset
bike_data_2020 = pd.concat([bike_jan, bike_feb, bike_mar, bike_apr, bike_may, bike_jun, bike_jul, bike_aug])
```

```python
bike_data_2020.head()
```

```python
bike_data_2020.tail()
```

```python
# Split the date and times
bike_data_2020[['StartDate','StartTime']] = bike_data_2020.starttime.str.split(expand=True)
bike_data_2020[['EndDate','EndTime']] = bike_data_2020.stoptime.str.split(expand=True)
```

```python
bike_data_2020.dtypes
```

```python
# Export the bike data to a csv, use encoding="utf-8
bike_data_2020.to_csv("data/bikedata2020.csv", encoding="utf-8", index=False)
```

### Now Cleanup the NYC COVID cases csv

```python
# COVID data downloaded from https://covidtracking.com/data/download from Jan 2020 to August 2020
covid_path = "data/new-york-history.csv"
```

```python
# Read in the COVID data
covid = pd.read_csv(covid_path)
```

```python
#Increase the number of columns visable to see which ones to drop
pd.set_option('display.max_columns', 50)
```

```python
covid.head()
```

```python
#Create a new dataframe with the required columns
covid2020 = covid[["date", "deathIncrease", "hospitalizedCurrently", "inIcuCurrently", "negative", "negativeIncrease", "positive", "positiveIncrease", "recovered", "totalTestResultsIncrease", ]]
```

```python
covid2020.head()
```

```python
# Export the COVID data to a csv, use encoding="utf-8
covid2020.to_csv("data/covid2020.csv", encoding="utf-8", index=False)
```
